### 1) Your business problem to solve:

We aim to provide the insurance company a proper prediction of the insurance claim for each customer and deeply understand the key risk factors to forecast the cost of the insurance claim so that the company can be able to know every detail about how to assess the claim circumstances and how much they should pay for the beneficiary, and thus the main goal is to maximize the profit of the insurance company based on the information of insurance price and the predicted claim.

### 2) Why the problem is important to solve?

Suppose a customer is devastated by a car accident, her/his focus is on the things that matter most: family, friends and other loved ones but pushing paper with the insurance agent is the last place, in this case, a claim prediction model can be extremely useful for the customer to see how much claim she/he can get with the advantage of saving a large amount of time.

### 3) What is your idea to solve the problem?

We have a data set containing 116 categorical variables and 14 continuous variables representing the information of the customers in Allstate Insurance Company, each column represents a claim from the company. However, the labels of these variables are hidden due to information security of the customers. Our idea to solve the problem is using neural network to build a model, training the model to get more precise predictions and predicting the claim for each customer.

### 4) What differences you could make with your data science approach?

In past, the trade-off between the processing duration after filing a claim to the insurance company and the benefit of customers has aroused much attention not only to insurance companies but also the policy holder. Fortunately, using neural network to predict the claim based on the information provided by the insurance company, we are able to efficiently and effectively help the insurance company to assess the circumstances and policy holder to estimate how much claim she/he can get roughly.

### 5) Why do you believe the idea deserves the investment of the "sharks"?

As mentioned above, the long duration between filing a claim and get money from the insurance company might be a big issue for the policy holder and beneficiary.

1. For the insurance company, getting the possible claim quickly can provide a useful reference to help them assess the circumstances for the policy;

2. For the customers, before filing a claim to the insurance company, they could have an expectation in advance with the model's fast calculation and prediction, which possibly avoids the communication gap between customers and companies and thus achieve a double-win between them;

3. For any companies, in long run, a model which enables them to deeply understand and analyze the data from customers can save much labor resources, reduce the attention on data but focus on the customers' need, and accordingly make more profits.  

Therefore, the idea deserves the investment of the "sharks" for companies in pursuit of improving their service quality to customers and for customers in pursuit of more credible and trustful information.
